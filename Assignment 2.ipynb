1. Show a breakdown of distance from home by job role and attrition.


```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Employee_info = pd.read_csv('Analysisdata.csv')
```

```python
Employee_info
```

```python
Employee_info.groupby('Attrition')['DistanceFromHome'].sum().reset_index()
```

```python
Employee_info.groupby('Attrition')['JobRole'].count().reset_index()
```

## Commentary


**Attrition v Distance from Home
- Overally Highest distance from home resulted in high levels of attrition ie the greater the distence from home, the higher the level of attrition
- Less distance from home has less impact on attrition levels

**Attrition v Job role
- 1233 members with different Job roles were not affected by attrition not affected by attrition where as 237 were affected by attrition

```python
Employee_info.groupby(['JobRole', 'Attrition'])['DistanceFromHome'].count().reset_index()
```

# Assignment 3 Q1 ANALYSIS

```python
sns.set(style="whitegrid", color_codes=True, rc={'figure.figsize':(8,4)})
#tips = sns.load_dataset("tips")
g = sns.lineplot(data=Employee_info, x="JobRole", y="DistanceFromHome", hue='Attrition')
#.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.xticks(rotation=90)
```
